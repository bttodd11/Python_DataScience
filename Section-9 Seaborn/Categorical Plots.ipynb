```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# Estimator is only counting the number of occurencesm you only set the xvalue and dataset
sns.countplot(x='sex', data=tips)
```

```python
# Box plot is used to show the distribution of quanitative data
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```
