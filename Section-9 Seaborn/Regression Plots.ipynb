```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','x'], scatter_kws={'s': 100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,col='day', hue='sex',aspect=0.6, size=8)
```
