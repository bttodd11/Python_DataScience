```python
import seaborn as sns
# This is needed to get charts to work in the notebook
%matplotlib inline 
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
# Annot make the value display in the heatmap
sns.heatmap(tc, annot=True)
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp,cmap='magma', linecolor='white', lw=.4)
```

```python
sns.clustermap(fp)
```
