```python
import matplotlib.pyplot as plt
```

```python
# Allows you to see the plots you create inside the Jupyter notebook 
%matplotlib inline
```

```python
import numpy as np
# Linspace is creating 11 points from 0-5
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# Use Plt.show() if you are not in a Jupyter Notebook
```

```python
#Subplot create 2 plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'-b')
```

```python
# OO Method
fig = plt.figure()

# Adding where I want the axes to be placed
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Calling the method from the add_axes then plots the values
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
#Left, bottom, width, height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.3,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("Larger Plot")
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)


axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("Title")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")

ax.plot(x, x ** 2,label="Y Squared")
ax.plot(x, x ** 3, label = "X Cubed")

ax.legend(loc=0)
```

```python
fig.savefig('my_pic.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.plot(x,y,color='#151750', linewidth=.5, linestyle='--', marker='o', markersize='10', markerfacecolor='white', markeredgewidth=2)
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
